```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

df=pd.read_csv("iris.csv")
print(df.describe)
df.head(5)
```

```python
X=df.iloc[:, [1, 2, 3, 4]].values #SepalLengthCm	SepalWidthCm	PetalLengthCm	PetalWidthCm
```

```python
from sklearn.cluster import KMeans
lst=list()

for _ in range(1,10):
    kmeans = KMeans(n_clusters = _, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    lst.append(kmeans.inertia_)
plt.plot(range(1, 10), lst)
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(X)
```

```python
%matplotlib notebook
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'r', label = 'Iris-setosa')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'b', label = 'Iris-versicolour')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'g', label = 'Iris-virginica')
```
